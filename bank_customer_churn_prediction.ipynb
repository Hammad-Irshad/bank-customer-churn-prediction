```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("/Churn_Modelling.csv")
df.sample(5)
```

```python
df.drop('CustomerId',axis='columns',inplace=True)
df.drop('Surname',axis='columns',inplace=True)
df.drop('RowNumber',axis='columns',inplace=True)
```

```python
def print_unique_col_values(df):
  for column in df:
    if df[column].dtypes=='object':
      print(f'{column}: {df[column].unique()}') 
```

```python
df1 = df['Gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df2 = pd.get_dummies(data=df, columns=['Geography','NumOfProducts'])
df2.columns
```

```python
df2.dtypes
```

```python
df2.sample(10)
```

```python
cols_to_scale = ['CreditScore','Tenure','Balance','EstimatedSalary']


from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
df2.sample(5)
```

```python
df2.dtypes
```

```python
#train test split for our ANN
X = df2.drop('Exited',axis='columns')
y = df2['Exited']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train[:10]
```

```python
len(X_train.columns)
```

```python
import tensorflow as tf
from tensorflow import keras



#or this will also work
model = keras.Sequential([
    keras.layers.Dense(15, input_shape=(15,), activation='relu'),
    keras.layers.Dense(1, activation='sigmoid') #ouput layer
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=120)
```

```python
yp = model.predict(X_test)
```

```python
yp[:20]
```

```python
#Convert probabilistic values to 0,1
y_pred = []
for element in yp:
  if element > 0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)
```

```python
y_pred[:20]
```

```python
#Evaluating the prediction
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```
